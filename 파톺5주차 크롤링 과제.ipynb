```python
crawl_url = ["https://news.v.daum.net/v/20201004215700006", "https://news.v.daum.net/v/20201020153505519", "https://news.v.daum.net/v/20201020153609574"]
```

```python
for i in crawl_url:
    # requests를 이용해 원하는 웹페이지의 소스코드를 불러옴
    res = requests.get(i)
    #print(res.content)
    #BeautifulSoup를 이용해 파싱
    title = bs(res.content, 'html.parser')
    print(title.select('#cSub > div > h3')[0].text)
```

```python
for i in crawl_url:
    # requests를 이용해 원하는 웹페이지의 소스코드를 불러옴
    res = requests.get(i)
    #print(res.content)
    #BeautifulSoup를 이용해 파싱
    company = bs(res.content, 'html.parser')
    print(company.select('#mArticle > div.foot_view > div.cp_view > strong')[0].text.split(' ')[0])
```

```python
df = pd.DataFrame(columns=['title', 'company', 'url'])
df
```

```python
a = []

for i in crawl_url:
    res = requests.get(i)
    title = bs(res.content, 'html.parser')
    a.append(title.select('#cSub > div > h3')[0].text)
    print(a)
```
